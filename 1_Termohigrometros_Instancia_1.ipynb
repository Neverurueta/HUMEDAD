```python
Coef = pd.read_excel('1.Termohgrometros_Instancia_1.xlsx', sheet_name='COE')
Incertidumbre= Mediciones_T.loc[:,['INDICACIÓN DE REFERENCIA','INDICACIÓN EBP']]
Incertidumbre["IT Calibración Patrón_1"] = T1["INCERTIDUMBRE TÍPICA"].max() * Coef.loc[0,'TEMPERATURA']
Incertidumbre["IT Calibración Patrón_2"] = T2["INCERTIDUMBRE TÍPICA"].max() * Coef.loc[1,'TEMPERATURA']
Incertidumbre["IT Deriva Patrón1"] = T1["DERIVA"].max() / (np.sqrt(3)) * Coef.loc[2,'TEMPERATURA']
Incertidumbre["IT Deriva Patrón2"] = T2["DERIVA"].max() / (np.sqrt(3)) * Coef.loc[3,'TEMPERATURA']
Incertidumbre["IT Res Patrón1"] = T1.loc[0,"RESOLUCIÓN"]/ (np.sqrt(12)) * Coef.loc[4,'TEMPERATURA']
Incertidumbre["IT Res Patrón2"] = (T2.reset_index().drop(["index"], axis=1)).loc[0,"RESOLUCIÓN"]/ (np.sqrt(12)) * Coef.loc[5,'TEMPERATURA']# con
#T2.reset_index().drop(["index"], axis=1)  se reinician los indices de T2
Incertidumbre["Interpolación1"] = np.sqrt((((np.sum(PT1R["RESIDUOS"].pow(2)))))/(T1.loc[0,'PARÁMETROS DEL POLINOMIO']-T1.loc[0,'GRADO DEL POLINOMIO'])) * Coef.loc[6,'TEMPERATURA']
Incertidumbre["Interpolación2"] = np.sqrt((((np.sum(PT2R["RESIDUOS"].pow(2)))))/((T2.reset_index().drop(["index"], axis=1)).loc[0,'PARÁMETROS DEL POLINOMIO']-(T2.reset_index().drop(["index"], axis=1)).loc[0,'GRADO DEL POLINOMIO'])) * Coef.loc[7,'TEMPERATURA']
Incertidumbre["IT Unifor Medio"] = (M1_T['UNIFORMIDAD'].max() / np.sqrt(3)) * Coef.loc[8,'TEMPERATURA']
Incertidumbre["IT Estab Medio"] = (M1_T['ESTABILIDAD'].max() / np.sqrt(3)) * Coef.loc[9,'TEMPERATURA']
Incertidumbre["Metodo ANOVA"] = 0.0471387117098546 * Coef.loc[10,'TEMPERATURA']
Incertidumbre["Incert del Sistema"] = np.sqrt(np.sum(Incertidumbre.iloc[:,2:].pow(2).T))*Coef.loc[11,'TEMPERATURA']
Incertidumbre["IT Resolución EBP"] = ((pd.read_excel('1.Termohgrometros_Instancia_1.xlsx', sheet_name='Temperatura').iloc[0,1])/np.sqrt(12)) * Coef.loc[12,'TEMPERATURA']
Incertidumbre["IT Histeresis"] = ((pd.read_excel('1.Termohgrometros_Instancia_1.xlsx', sheet_name='HistTemp').iloc[2:4,2].mean() - pd.read_excel('1.Termohgrometros_Instancia_1.xlsx', sheet_name='HistTemp').iloc[5:,2].mean())/np.sqrt(12))* Coef.loc[13,'TEMPERATURA']
Incertidumbre["Incert de la Corrección"] = np.sqrt(np.sum(Incertidumbre.iloc[:,13:].pow(2).T))
Incertidumbre.reset_index(inplace=True, drop=True)
```
